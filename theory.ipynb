Three principles:
- **heterogeneity**: the information present often shows diverse qualities, structures, and representations
- **connections**:  they are often related and share commonalities
- **interactions**: give rise to **new information** when used for **task inference**

Six challenges:
- **representation**: how to summarize multimodal data to reflect the ***heterogeneity*** and ***interconnections*** between individual modality elements
  - ***representation fusion***: integrating information from two or more modalities to capture cross-modal interactions
  - ***representation coordination***: interchanging cross-modal information to keep the same number of representations but improve multimodal contextualization
  - ***representation fission***: creating a larger set of disjoint representations that reflects knowledge about internal structure such as data clustering or factorization
- **alignment**: captures the ***connections*** and ***interactions*** between multiple local elements according to their structure
  - ***discrete alignment***: identifying connections between discrete elements across modalities
  - ***continuous alignment***: modeling alignment between continuous modality signals with ambiguous segmentation
  - ***contextualized representations***: learning better representations by capturing cross-modal interactions between elements
- **reasoning**: combine the information from multimodal evidence in a principled way that respects the structure of the problem to give more robust and interpretable predictions
- **generation**:  generative process to produce raw modalities that reflect cross-modal interactions, structure, and coherence
- **transference**:  transfer information from high-resource modalities to low-resource ones and their representations
- **quantification**: revisits the previous challenges to give deeper empirical and theoretical understanding of modality heterogeneity, interconnections, and the learning process
